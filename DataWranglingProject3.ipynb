## Treatment of missing values 



#### Mean Impportation

- - Mean tmputation , replace the missing values with the variable mean

```python
from sklearn.impute import SimpleImputer
mean_imputer= SimpleImputer(missing_values=np.nan,strategy='mean')
mean_imputer= mean_imputer.fit(classg)
imputed_classg=mean_imputer.transform(classg.values)
df1=pd.DataFrame(data=imputed_classg,columns=classg.columns)

df1
```

```python
df1.isna().any()
```
