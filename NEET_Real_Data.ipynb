#### Calculating Confidence Interval

```python
mean_score = np.mean(df['marks']).round(2)
std_score = np.std(df['marks'])
n = len(df)
confidence_level = 0.95

# Let's calculate the z-score for the given confidence interval
z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)


# Standard Error(SE) = (sigma / root(n))
standard_error = std_score / np.sqrt(n)


# Margin Of Error(ME)
margin_of_error = np.round(z_score * standard_error, 2)


# Confidence Interval
confidence_interval = (mean_score - margin_of_error, mean_score + margin_of_error)


print(f"95% Confidence Interval: {confidence_interval}")
```

### Observations:

- We are 95% confident that the true average exam score for all students who appeared for the exam lies between 216.95 and 217.37

```python
import pandas as pd
df=pd.read_csv('all_data.csv')
```

```python

aggregated_df = df.groupby(['center_state', 'center_name', 'center_id', 'center_city', 'serial_no'], as_index=False).agg({
    'score': 'sum'  # Aggregating scores, use 'sum' to combine them
})

# Define the mapping dictionary for state names
state_mapping = {
    'tamil nadu': 'Tamil Nadu',
    'tamilnadu': 'Tamil Nadu',
    'andhra pradesh': 'Andhra Pradesh',
    'andhrapradesh': 'Andhra Pradesh',
    'madhya pradesh': 'Madhya Pradesh',
    'madhyapradesh': 'Madhya Pradesh',
    'west bengal': 'West Bengal',
    'westbengal': 'West Bengal',
    'himachal pradesh': 'Himachal Pradesh',
    'himachalpradesh': 'Himachal Pradesh',
    'uttar pradesh': 'Uttar Pradesh',
    'uttarpradesh': 'Uttar Pradesh',
    'jammu & kashmir': 'Jammu & Kashmir',
    'jammu &kashmir': 'Jammu & Kashmir',
    'jammu& kashmir': 'Jammu & Kashmir',
    'arunachal pradesh': 'Arunachal Pradesh',
    'arunachalpradesh': 'Arunachal Pradesh',
    'puducherry(ut)': 'Puducherry (UT)',
    'puducherry (ut)': 'Puducherry (UT)',
    'dadra & nagar haveli (ut)': 'Dadra & Nagar Haveli (UT)',
    'dadra & nagarhaveli (ut)': 'Dadra & Nagar Haveli (UT)',
    'daman & diu(ut)': 'Daman & Diu (UT)',
    'daman & diu (ut)': 'Daman & Diu (UT)'
}

# Normalize and map state names
aggregated_df['center_state'] = aggregated_df['center_state'].str.lower().str.strip().replace(state_mapping)

# Print the cleaned DataFrame
print(aggregated_df.head())

# Optionally, save the aggregated data to a new file
# aggregated_df.to_csv('aggregated_data.csv', index=False)
```

```python
# Filter for centers in Bihar
bihar_centers = aggregated_df[aggregated_df['center_state'] == 'bihar']

# Get unique center names in Bihar
unique_center_names_bihar = bihar_centers['center_name'].unique()

# Print the unique center names
print('Unique center names in Bihar:')
for center_name in unique_center_names_bihar:
    print(center_name)

```

```python
center_name_target = 'A N COLLEGE, BOARING RAOD, SHRI KRISHNAPURI, PATNA, BIHAR'
state_name_target = 'bihar'

filtered_df = aggregated_df[(aggregated_df['center_state'] == state_name_target) & (aggregated_df['center_name'] == center_name_target)]

high_scores_filtered = filtered_df[filtered_df['score'] >= 650]

num_students_high_scores = high_scores_filtered.shape[0]

print(f'Number of students who scored more than 600 in "{center_name_target}": {num_students_high_scores}')

```

```python


# Define the target center name and state
target_center_name = 'HOLY MISSION SECONDARY SCHOOL, ADAMPUR WEST OF KHAGAUL ROAD, PATNA, BIHAR'
state_name = 'bihar'

# Filter and count the number of students scoring more than 650
num_students_high_scores = aggregated_df[
    (aggregated_df['center_state'] == state_name) &
    (aggregated_df['center_name'] == target_center_name) &
    (aggregated_df['score'] > 650)
].shape[0]

num_students_700_score = aggregated_df[
    (aggregated_df['center_state'] == state_name) &
    (aggregated_df['center_name'] == target_center_name) &
    (aggregated_df['score'] >= 700)
].shape[0]

# Print the result
print(f'Number of students who scored more than 650 in "{target_center_name}": {num_students_high_scores}')
print(f'Number of students who scored 700 in "{target_center_name}": {num_students_700_score}')

```
